```python
dheko="https://www.cardekho.com/used-cars+in+mumbai?utm_campaign=&utm_device=c&utm_term=&network=g&utm_medium=cpc&utm_source=google&agid=130757335183&ap=&aoi=9302386&ci=15845178844&cre=573963134111&fid=&lop=1007776&ma=&mo=&pl=&ti=dsa-1588904871118&gclid=EAIaIQobChMInKaDw7m69QIVp5lmAh2e3AcJEAAYAyAAEgKIUvD_BwE"
```

```python
#connecting to webdriver
driver=webdriver.Chrome(r'C:\Program Files (x86)\chromedriver.exe')
#opening the url
driver.get(dheko)
time.sleep(3)
driver.maximize_window()
```

```python
#url
url_dheko=[]
for ud in driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a"):
    url_dheko.append(ud.get_attribute("href"))
```

```python
len(url_dheko)
```

```python
#creating empty list
brand_dheko=[]
model_dheko=[]
year_dheko=[]
variant_dheko=[]
fuel_dheko=[]
dkms_dheko=[]
transmission_dheko=[]
loc_dheko=[]
price_dheko=[]
#owner=[]
#insurance=[]
#service=[]
#tyres_replaced=[]

#for cu in url_c:
    #try:
        #driver.get(cu)
    #except WebDriverException as w:
        #print("exception",w)
#brand    
try:
    brd=driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a")
    for bc in brd:
        brand_dheko.append(bc.text.split(' ',maxsplit=2)[1])
except:
    brand_dheko.append('-')
    
#model
try:
    mode_c=driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a")
    for mc in mode_c:
        model_dheko.append(mc.text.split(' ',maxsplit=3)[2])
except NoSuchElemenrException as n:
    print("exception",n)
    
#year of manufacture
try:
    year_d=driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a")
    for yc in year_d:
        year_dheko.append(yc.text.split(' ',maxsplit=1)[0])
except:
    year_dheko.append('-')
    
#variant
try:
    version_c=driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/div[1]")
    for vc in version_c:
        variant_dheko.append(vc.text)
except:
    variant_dheko.append('-')
    
#fuel
try:
    fu_c=driver.find_elements_by_xpath("//div[@class='truncate dotlist']/span[2]")
    for fc in fu_c:
        fuel_dheko.append(fc.text)
except:
    fuel_dheko.append('-')
    
#driven kilometers
try:
    driven_c=driver.find_elements_by_xpath("//div[@class='truncate dotlist']/span[1]")
    for dc in driven_c:
        dkms_dheko.append(dc.text)
except:
    dkms_dheko.append('-')
        
#transmission type
try:
    transmission_c=driver.find_elements_by_xpath("//div[@class='truncate dotlist']/span[3]")
    for tc in transmission_c:
        transmission_dheko.append(tc.text)
except:
    transmission_dheko.append('-')
    
#price
try:
    rate_c=driver.find_elements_by_xpath("//span[@class='amnt ']")
    for rc in rate_c:
        price_dheko.append(rc.text)
except:
    price_dheko.append('-')
    
#location
try:
    loc_c=driver.find_elements_by_xpath("//div[@class='cityIcon']")
    for lc in loc_c:
        loc_dheko.append(lc.text)
except NoSuchElementException as m:
    print("exception",m)
```

```python
owner_dheko=[]
for du in url_dheko:
    try:
        driver.get(du)
    except WebDriverException as w:
        print("exception",w)
    #owner
    try:
        own=driver.find_element_by_xpath("/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[3]/div/div/div/div[1]/ul/li[4]/div/div")
        owner_dheko.append(own.text) 
    except:
        owner_dheko.append('-')
```
